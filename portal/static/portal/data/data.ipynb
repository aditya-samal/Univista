```python
import pandas as pd
```

```python
files = ['2016-1.csv','2016-2.csv','2016-3.csv','2016-4.csv','2016-5.csv','2016-6.csv',
         '2017-1.csv','2017-2.csv','2017-3.csv','2017-4.csv','2017-5.csv','2017-6.csv','2017-7.csv',
         '2018-1.csv','2018-2.csv','2018-3.csv','2018-4.csv','2018-5.csv','2018-6.csv','2018-7.csv',
         '2019-1.csv','2019-2.csv','2019-3.csv','2019-4.csv','2019-5.csv','2019-6.csv','2019-7.csv',
         '2020-1.csv','2020-2.csv','2020-3.csv','2020-4.csv','2020-5.csv','2020-6.csv',
         '2021-1.csv','2021-2.csv','2021-3.csv','2021-4.csv','2021-5.csv','2021-6.csv',
         '2022-1.csv','2022-2.csv','2022-3.csv','2022-4.csv','2022-5.csv','2022-6.csv',
         '2023-1.csv','2023-2.csv','2023-3.csv','2023-4.csv','2023-5.csv','2023-6.csv',
         ]
```

```python
all_dfs = []
for filename in files:
    new_df = pd.read_csv(filename)
    all_dfs.append(new_df)
    
df = pd.concat(all_dfs)
df.to_csv(path_or_buf= 'File.csv', index=False)
```

```python
data= pd.read_csv('File.csv')
```

```python
def data_clean(df):  
    def isint(x):
        try:
            int(x)
            return True
        except:
            return False
        
    for i in df.index:
        if isint(df.loc[i, 'Opening Rank']):
            df.loc[i, 'Opening Rank'] = int(df.loc[i, 'Opening Rank'])
        else:
            df.loc[i, 'Opening Rank'] = 0    
        
    for i in df.index:
        if isint(df.loc[i, 'Closing Rank']):
            df.loc[i, 'Closing Rank'] = int(df.loc[i, 'Closing Rank'])
        else:
            df.loc[i, 'Closing Rank'] = 0
    
    df['Opening Rank']=df['Opening Rank'].astype('int64')
    df['Closing Rank']=df['Closing Rank'].astype('int64')
    df['Institute']=df['Institute'].astype('category')
    df['Academic Program Name']=df['Academic Program Name'].astype('category')
    df['Quota']=df['Quota'].astype('category')
    df['Seat Type']=df['Seat Type'].astype('category')
    df['Gender']=df['Gender'].astype('category')
    print(df.info())

data_clean(data)
```
